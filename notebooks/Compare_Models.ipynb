# Model Comparison

This notebook compares the performance of different sentiment classification models on the UIT-VSFC dataset.

Models:
- TF-IDF + Logistic Regression
- PhoBERT

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.DataFrame({
    "Model": [
        "TF-IDF + Logistic Regression",
        "PhoBERT"
    ],
    "Accuracy": [
        0.82,
        0.8802
    ],
    "F1": [
        0.82,
        0.8800
    ]
})
```

```python
plt.figure(figsize=(10,5))

plt.bar(
    results["Model"],
    results["Accuracy"]
)

plt.title("Model Accuracy Comparison")

plt.ylabel("Accuracy")

plt.xticks(rotation=20)

plt.tight_layout()

plt.savefig(
    "../images/model_accuracy_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.bar(
    results["Model"],
    results["F1"]
)

plt.title("Model F1 Score Comparison")

plt.ylabel("F1 Score")

plt.xticks(rotation=20)

plt.tight_layout()

plt.savefig(
    "../images/model_f1_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
baseline_acc = 0.82

best_acc = 0.8802

improvement = (
    (best_acc - baseline_acc)
    / baseline_acc
) * 100

print(
    f"Accuracy improvement: {improvement:.2f}%"
)
```

## Conclusion

The PhoBERT-based model significantly outperformed the traditional TF-IDF + Logistic Regression baseline.

Accuracy improved from 82.0% to 88.0%, corresponding to a relative improvement of 7.34%.
